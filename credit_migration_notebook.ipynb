```python
def get_security_name(ticker):
    # Extract the ticker symbol before the space
    symbol = ticker.split(' ')[0]
    symbol = symbol.replace('/', '.')
    ticker_obj = yf.Ticker(symbol)
    return ticker_obj.info.get('longName', 'Name not found')

# Apply the function to each ticker in the 'Security' column
resampled_ratings_data['Security Name'] = resampled_ratings_data['Security'].apply(get_security_name)
```

## Parsing Adverse News

```python


# Function to convert relative date to 'yyyy-mm-dd' format
def convert_relative_date(date_str):
    # Extracting the first two characters from the date string
    days_ago = int(date_str[:2])
    # Calculating the date by subtracting the number of days from the current date
    calculated_date = datetime.now() - timedelta(days=days_ago)
    # Formatting the date in 'yyyy-mm-dd' format
    formatted_date = calculated_date.strftime( '%b %d, %Y')
    return formatted_date

# Function to convert date string to datetime object
def convert_to_datetime(date_str):
    try:
        return datetime.strptime(date_str,  '%b %d, %Y')
    except ValueError:
        return None


def google_search_to_dataframe(company_name, evaluation_date , api_key = ''):
    # note that the API key has been removed

    cx = ###Masked

    # Create service object
    service = build('customsearch', 'v1', developerKey=api_key)

    one_year_ago = pd.to_datetime(evaluation_date) - timedelta(days=365)
    formatted_one_year_ago = one_year_ago.strftime('%Y-%m-%d')


    # Construct the search query with date restriction
    search_query = f'"{company_name}" ("v" OR "vs" OR "plaintiff" OR "defendant" OR "lawsuit" OR "litigation" OR "legal action" OR "legal dispute" OR "court case" OR "legal proceedings" OR "legal matter" OR "legal challenge" OR "jurisdiction" OR "attorney" OR "fraud" OR "custody")'


    # Add date restriction to the query
    search_query_with_date = f'{search_query} daterange:{formatted_one_year_ago}-'


    res = service.cse().list(
        q=search_query_with_date,
        cx=cx,
        num = 10  # Limit to top 10 results
    ).execute()

    # Extracting necessary information from search results
    search_results = []
    # print(res)
    for item in res['items']:
        title = item['title']
        snippet = item['snippet']
        link = item['link']
        date_str = item['snippet'][:12]
        if 'days ago' in date_str:
            date_str = convert_relative_date(date_str)

        # Convert date string to datetime object
        if (date_str is not None):
            date = convert_to_datetime(date_str)
        if date is not None and ((evaluation_date - date).days <= 365) and ((evaluation_date - date).days > 0):
            search_results.append({'company':company_name, 'title': title, 'snippet': snippet, 'link': link, 'report_date': date})

    # Sort the search results based on date in descending order
    sorted_results = sorted(search_results, key=lambda x: x['report_date'], reverse=True)

    # Create DataFrame from search results
    df = pd.DataFrame(sorted_results)

    return df

```

```python
responses = []
for row in tqdm(df_news_deduped.iterrows()):
    new_snippet_1 = str(row[1]['company'])+ '  ' + str(row[1]['title']) + '  ' + str(row[1]['snippet'])
    response1 = pipe(new_snippet_1)

    new_snippet_2 = str(row[1]['title']) 
    response2 = pipe(new_snippet_2)

    new_snippet_3 = str(row[1]['snippet']) 
    response3 = pipe(new_snippet_3)

    responses.append([response1, response2, response3])
```

```python
combo_sentiment = []
title_sentiment = []
snippet_sentiment = []

combo_sentiment_score = []
title_sentiment_score = []
snippet_sentiment_score = []

for response in responses:
    combo_sentiment.append( response[0][0]['label'])
    combo_sentiment_score.append( response[0][0]['score'])

    title_sentiment.append( response[1][0]['label'])
    title_sentiment_score.append( response[1][0]['score'])

    snippet_sentiment.append( response[2][0]['label'])
    snippet_sentiment_score.append( response[2][0]['score'])
```

```python
# creating positive and negative migration indicators
rating_map = {
    'AAA': 1, 
    'AA+': 2, 
    'AA': 3, 
    'AA-': 4, 
    'AA *-': 5, 
    'AA *': 6, 
    'A+': 7, 
    'A': 8, 
    'A-': 9, 
    'A- *-': 10,
    'BBB+': 11, 
    'BBB *+': 12, 
    'BBB': 13, 
    'BBB-': 14, 
    'BBB- *+': 15,
    'BB+': 16, 
    'BB': 17, 
    'BB-': 18,
    'B+': 19, 
    'B': 20, 
    'B-': 21
}



merged_news_filtered['BASE_RATING_NUM'] = merged_news_filtered['BASE_RATING'].map(rating_map)
merged_news_filtered['RATING_NUM'] = merged_news_filtered['rating'].map(rating_map)

# Create the direct_mig_ind column
merged_news_filtered['direct_mig_ind'] = (merged_news_filtered['RATING_NUM'] - merged_news_filtered['BASE_RATING_NUM']).apply(
    lambda x: 1 if x < 0 else (-1 if x > 0 else 0)
)
```

```python


# Companies where no articles / materials were found
merged_news_filtered['ordinal_label_union'] = merged_news_filtered['ordinal_label_union'].fillna(-9)

ordinal_summary = merged_news_filtered.groupby('ordinal_label_union').agg({'Migration_ind':['sum','count']}).reset_index()
ordinal_summary['mig_perc']  = ordinal_summary['Migration_ind']['sum'] / ordinal_summary['Migration_ind']['count']
ordinal_summary[ordinal_summary['ordinal_label_union']!=-9][['ordinal_label_union','mig_perc']]
```

```python
unstacked_df = \
    merged_news_filtered.groupby(['ordinal_label_union','direct_mig_ind'])\
            .agg({'direct_mig_ind':'count'}).unstack().fillna(0)
        # [merged_news_filtered['ordinal_label_union']!=-9]\
        

row_sums = unstacked_df.sum(axis=1)
proportions_df = unstacked_df.div(row_sums, axis=0)

# Rename columns for clarity
proportions_df.columns = proportions_df.columns.droplevel()
(proportions_df*100).round(2)
```

#### Observations from above table
##### 1. Any kind of public news does not lead to improvement in credit ratings
##### 2. None of the news items marked positive by the FinBERT does through a credit rating reduction
##### 3. There is a slight difference in chances of a credit rating reduction of a negative news is encountered. However, more data is required to be able to comment whether these findings are statistically significant
##### 4. There is definitely factors beyond the public news captured in this analysis that impacts the credit ratings migrations, as evident from the first row
